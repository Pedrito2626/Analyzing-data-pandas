# Distribución de Emisiones de CO2 por Marca
En esta gráfica de nos muestra el rango en el que se encuentran las emisiones de CO2 por marca.

Marcas como chevrolet, ford, Mercedes, AUDI y BMW tienen un rango de emisiones generalmente bajo.

Toyota es la amrca que en general genera menos emisiones de CO2.

Sin embargo, la que en general genera más emisiones de CO2 es GMC.

A pesar de lo anterior, ford y chevrolet tiene registros donde generan más emisiones que cualquier otro.
```python

````python
plt.figure(figsize=(14,8))
top_brands = df['Make'].value_counts().head(8).index
df_top_brands = df[df['Make'].isin(top_brands)]

sns.boxplot(data=df_top_brands, x='Make', y='CO2 Emissions(g/km)', palette='viridis')
plt.title('Distribución de Emisiones de CO2 por Marca', fontsize=16)
plt.xlabel('Marca', fontsize=12)
plt.ylabel('Emisiones CO2 (g/km)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

# Relación entre tamaño del motor y emisiones
En esta gráfica podemos ver claramente una relación lineal entre el tamaño del motor y las emisiones de CO2.

A medida que el tamaño del motor aumenta, las emisiones de CO2 también tienden a aumentar. 

Sin necesidad  de usar la matriz de correlación podemos intuir que su relación es muy cercana a 1.

Los vehículos que usan el combustible tipo Z tienden a tener un tamaño de motor más grande y, por lo tanto, mayores emisiones de CO2.

Mientras que los vehículos que usan combustible tipo X tienden a todo lo contrario, un tamaño de motor más pequeño y menores emisiones de CO2.
```python

````python
# Relación entre tamaño del motor y emisiones
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Engine Size(L)', y='CO2 Emissions(g/km)', hue='Fuel Type', palette='magma', alpha=0.7)
plt.title('Relación entre Tamaño del Motor y Emisiones de CO2', fontsize=16)
plt.xlabel('Tamaño del Motor (L)', fontsize=12)
plt.ylabel('Emisiones CO2 (g/km)', fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
````

# Matriz de correlación y mapa de calor
En esta gráfica se nos muestra la matriz de correlación entre las variables numéricas del dataset.

Como se esperaba del ejemplo anterior, el tamaño del motor y las emisiones de CO2 tienen una correlación muy alta (0.85).

Sin contar fuel consumption comb en (mpg), que es la única que no tiene una correlación alta con las demás variables, esto probablemente porque se evalua de manera diferente a las demás. Esta es medida en millas por galón, mientras que las demás están en litros por 100 km.

El resto prsenta una correlación alta entre sí, indicando la alta sincronización entre ellas.
```python

````python
# Seleccionar solo variables numéricas
numeric_cols = df.describe().columns

plt.figure(figsize=(10,8))
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Matriz de Correlación - Variables del Dataset', fontsize=16)
plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='Fuel Consumption City (L/100 km)', y='CO2 Emissions(g/km)', hue='Fuel Type', palette='magma', alpha=0.7)
plt.title('Relación entre Consumo de Combustible en Ciudad y Emisiones de CO2', fontsize=16)
plt.xlabel('Consumo de Combustible en Ciudad (L/100 km)', fontsize=12)
plt.ylabel('Emisiones CO2 (g/km)', fontsize=12)
plt.legend(title='Tipo de Combustible', loc='best', fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
````

Existe una alta relación entre las emisiones de CO2 y el consumo de combustible en ciudad (0.92 en la matriz de correlación), lo que indica que a mayor consumo de combustible, mayores emisiones de CO2.

Y, en la gráfic aanterior podemos apreciar la linealidad entre en sonsumo de combustible en ciudad y las emisiones de CO2, apreciando visualmente la alta correlación entre estas dos variables.

Además, podemos ver la diferencia según el tipo de combustible, donde todos siguen cumpliendo la misma tendencia, pero con diferentes niveles de emisiones. El combustible E es el que menos emisiones de CO2 general a pesar de que es el que más combustible consume en ciudad.

````python
nba = pd.read_csv(
    "https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv",
    index_col="Unnamed: 0",
)
nba
````

````python
nba["College"] = nba["College"].fillna("No College")
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
nba = nba.drop(457)  # ? elimina la fila vacía
nba.describe()  # ? Muestra estadísticas descriptivas de las columnas numéricas
````

````python
Top5equiposMasSalarios = (
    nba.groupby("Team")["Salary"].sum().sort_values(ascending=False).head(5)
)
Top5equiposMasSalarios
````

````python
nba["Age-Group"] = pd.cut(
    nba["Age"], bins=[0, 25, 35, 40], labels=["<25", "25-35", ">35"]
)
fig, ax = plt.subplots(figsize=(10, 6))
sns.countplot(
    data=nba, x="Age-Group", hue="Position", palette="pastel", edgecolor="black", ax=ax
)
ax.set_title("Número de jugadores por grupo de edad y posición")
ax.set_xlabel("Grupo de Edad")
ax.set_ylabel("Número de Jugadores")
````

````python
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
sns.boxplot(
    data=nba, x="Position", y="Salary", palette="pastel", ax=axes[0], hue="Position"
)
axes[0].set_title("Salarios por Posición")
axes[0].set_xlabel("Posición")
axes[0].set_ylabel("Salario")

sns.boxplot(
    data=nba, x="Age-Group", y="Salary", palette="pastel", ax=axes[1], hue="Age-Group"
)
axes[1].set_title("Salarios por Grupo de Edad")
axes[1].set_xlabel("Grupo de Edad")
axes[1].set_ylabel("")
plt.tight_layout()
plt.show()
````

````python
# ? Construir una tabla que muestre el salario promedio por equipo y por posición para los cinco equipos con mayor salario. Visualizar los resultados en un mapa de calor.
avg_salary = nba.groupby(["Team", "Position"])["Salary"].mean().unstack()
avg_salary.fillna(0, inplace=True)  # ? Rellena los valores NaN con 0
fig, ax = plt.subplots(figsize=(12, 8))

sns.heatmap(avg_salary, annot=True, fmt=".0f", cmap="YlGnBu", ax=ax)
ax.set_title("Salario Promedio por Equipo y Posición")
ax.set_xlabel("Posición")
ax.set_ylabel("Equipo")
plt.show()
````
